```python
#Oversampling
from sklearn.model_selection import cross_val_score, StratifiedKFold
from imblearn.pipeline import Pipeline as imbpipeline
from imblearn.over_sampling import SMOTE
```

```python
arbol = DecisionTreeClassifier(max_depth = 3)
pipeline = imbpipeline([('oversample', SMOTE()), ('arbol', arbol)])
skf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 5)
resultado_arbol = cross_val_score(pipeline, x, y, cv = skf, scoring = 'f1')
print(f'F1 (Decision Tree):{resultado_arbol.mean()}')
```

```python
random_forest = RandomForestClassifier(max_depth = 2)
pipeline = imbpipeline([('oversample', SMOTE()), ('random_forest', random_forest)])
skf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 5)
resultado_rf = cross_val_score(pipeline, x, y, cv = skf, scoring = 'f1')
print(f'F1 (Random Forest):{resultado_rf.mean()}')
```

```python
#Undersampling
from imblearn.pipeline import Pipeline as imbpipeline
from imblearn.under_sampling import NearMiss
```

```python
arbol = DecisionTreeClassifier(max_depth = 3)
pipeline = imbpipeline([('undersample', NearMiss(version = 3)), ('arbol', arbol)])
skf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 5)
resultado_arbol = cross_val_score(pipeline, x, y, cv = skf, scoring = 'f1')
print(f'F1 (Decision Tree):{resultado_arbol.mean()}')
```

```python
random_forest = RandomForestClassifier(max_depth = 2)
pipeline = imbpipeline([('undersample', NearMiss(version = 3)), ('random_forest', random_forest)])
skf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 5)
resultado_rf = cross_val_score(pipeline, x, y, cv = skf, scoring = 'f1')
print(f'F1 (Random Forest):{resultado_rf.mean()}')
```

```python
#Elige el modelo de Ã¡rbol obtuvo el mejor rendimiento
#Por lo tanto, vamos a balancear los datos y utilizar este modelo en todos los datos balanceados. Por Ãºltimo, vamos a validar el modelo en datos de prueba, no vistos hasta ahora:
from imblearn.under_sampling import NearMiss
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split

# Define variables x and y
x = datos.drop('Churn', axis = 1)
y = datos['Churn']

# Split data into training, validation (25%), and test (5%)
x, x_prueba, y, y_prueba = train_test_split(x, y, stratify = y, test_size = 0.05, random_state = 5)
x_entrenamiento, x_val, y_entrenamiento, y_val = train_test_split(x, y, stratify = y, random_state = 5)

# Apply one-hot encoding to x and x_prueba
x = one_hot.fit_transform(x)
x_prueba = one_hot.transform(x_prueba)

undersample = NearMiss(version = 3)
x_balanceado, y_balanceado = undersample.fit_resample(x, y)

modelo = RandomForestClassifier(max_depth = 2)
modelo.fit(x_balanceado, y_balanceado)
y_previsto = modelo.predict(x_prueba)

print(classification_report(y_prueba, y_previsto))
ConfusionMatrixDisplay.from_predictions(y_prueba, y_previsto);
```

#ConclusiÃ³n

Modelo mÃ¡s adecuado: **Ãrbol de DecisiÃ³n**

Aunque el Random Forest muestra una alta precisiÃ³n (0.75), su recall (0.051) es extremadamente bajo. Esto significa que identifica correctamente muy pocos clientes que realmente hacen churn, aunque cuando predice churn, suele acertar.

En cambio, el Ãrbol de DecisiÃ³n tiene:

* Mejor exactitud general (0.77 vs 0.75),

* Recall mucho mÃ¡s alto (0.458),

* F1-Score mÃ¡s equilibrado (0.506 vs 0.096), lo que indica un mejor balance entre precisiÃ³n y recall.

Dado que en churn lo mÃ¡s importante es identificar a los clientes que efectivamente se fugarÃ¡n, el recall es la mÃ©trica mÃ¡s crÃ­tica. En este caso, el Ã¡rbol de decisiÃ³n es claramente superior.

**Variables claves**:
Las variables 'customer.tenure' (antigÃ¼edad del cliente) y 'account.Charges.Total' (monto total pagado) son las mÃ¡s relevantes para explicar la fuga de clientes. Esto sugiere que:

* Clientes con poca antigÃ¼edad tienen mayor riesgo de churn.

* Clientes que han pagado menos (o poco tiempo) tambiÃ©n tienen mayor probabilidad de abandonar el servicio.

El modelo Ãrbol de DecisiÃ³n entrega el mejor balance entre identificar correctamente a los clientes que se fugan y evitar falsos positivos. Por lo tanto, es el modelo mÃ¡s adecuado para implementar.
