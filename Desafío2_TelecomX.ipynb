```python
#Oversampling
from sklearn.model_selection import cross_val_score, StratifiedKFold
from imblearn.pipeline import Pipeline as imbpipeline
from imblearn.over_sampling import SMOTE
```

```python
arbol = DecisionTreeClassifier(max_depth = 3)
pipeline = imbpipeline([('oversample', SMOTE()), ('arbol', arbol)])
skf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 5)
resultado_arbol = cross_val_score(pipeline, x, y, cv = skf, scoring = 'f1')
print(f'F1 (Decision Tree):{resultado_arbol.mean()}')
```

```python
random_forest = RandomForestClassifier(max_depth = 2)
pipeline = imbpipeline([('oversample', SMOTE()), ('random_forest', random_forest)])
skf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 5)
resultado_rf = cross_val_score(pipeline, x, y, cv = skf, scoring = 'f1')
print(f'F1 (Random Forest):{resultado_rf.mean()}')
```

```python
#Undersampling
from imblearn.pipeline import Pipeline as imbpipeline
from imblearn.under_sampling import NearMiss
```

```python
arbol = DecisionTreeClassifier(max_depth = 3)
pipeline = imbpipeline([('undersample', NearMiss(version = 3)), ('arbol', arbol)])
skf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 5)
resultado_arbol = cross_val_score(pipeline, x, y, cv = skf, scoring = 'f1')
print(f'F1 (Decision Tree):{resultado_arbol.mean()}')
```

```python
random_forest = RandomForestClassifier(max_depth = 2)
pipeline = imbpipeline([('undersample', NearMiss(version = 3)), ('random_forest', random_forest)])
skf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 5)
resultado_rf = cross_val_score(pipeline, x, y, cv = skf, scoring = 'f1')
print(f'F1 (Random Forest):{resultado_rf.mean()}')
```

```python
#Elige el modelo de árbol obtuvo el mejor rendimiento
#Por lo tanto, vamos a balancear los datos y utilizar este modelo en todos los datos balanceados. Por último, vamos a validar el modelo en datos de prueba, no vistos hasta ahora:
from imblearn.under_sampling import NearMiss
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split

# Define variables x and y
x = datos.drop('Churn', axis = 1)
y = datos['Churn']

# Split data into training, validation (25%), and test (5%)
x, x_prueba, y, y_prueba = train_test_split(x, y, stratify = y, test_size = 0.05, random_state = 5)
x_entrenamiento, x_val, y_entrenamiento, y_val = train_test_split(x, y, stratify = y, random_state = 5)

# Apply one-hot encoding to x and x_prueba
x = one_hot.fit_transform(x)
x_prueba = one_hot.transform(x_prueba)

undersample = NearMiss(version = 3)
x_balanceado, y_balanceado = undersample.fit_resample(x, y)

modelo = RandomForestClassifier(max_depth = 2)
modelo.fit(x_balanceado, y_balanceado)
y_previsto = modelo.predict(x_prueba)

print(classification_report(y_prueba, y_previsto))
ConfusionMatrixDisplay.from_predictions(y_prueba, y_previsto);
```

#Conclusión

Modelo más adecuado: **Árbol de Decisión**

Aunque el Random Forest muestra una alta precisión (0.75), su recall (0.051) es extremadamente bajo. Esto significa que identifica correctamente muy pocos clientes que realmente hacen churn, aunque cuando predice churn, suele acertar.

En cambio, el Árbol de Decisión tiene:

* Mejor exactitud general (0.77 vs 0.75),

* Recall mucho más alto (0.458),

* F1-Score más equilibrado (0.506 vs 0.096), lo que indica un mejor balance entre precisión y recall.

Dado que en churn lo más importante es identificar a los clientes que efectivamente se fugarán, el recall es la métrica más crítica. En este caso, el árbol de decisión es claramente superior.

**Variables claves**:
Las variables 'customer.tenure' (antigüedad del cliente) y 'account.Charges.Total' (monto total pagado) son las más relevantes para explicar la fuga de clientes. Esto sugiere que:

* Clientes con poca antigüedad tienen mayor riesgo de churn.

* Clientes que han pagado menos (o poco tiempo) también tienen mayor probabilidad de abandonar el servicio.

El modelo Árbol de Decisión entrega el mejor balance entre identificar correctamente a los clientes que se fugan y evitar falsos positivos. Por lo tanto, es el modelo más adecuado para implementar.

##Estrategia de retención
El modelo puede utilizarse para identificar perfiles de alto riesgo y anticipar la fuga de clientes, permitiendo diseñar estrategias de retención proactivas, especialmente enfocadas en:

* Nuevos clientes

* Clientes con baja facturación total

* Segmentos con baja fidelización histórica

En particular se propone fortalecer el vínculo temprano con el cliente y aumentar su percepción de valor desde los primeros meses con campañas de fidelización, onboarding, programa de rewards, descuentos por permanencia.
